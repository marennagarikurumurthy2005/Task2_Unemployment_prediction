```python
# Assuming unemployment_data is already defined as a DataFrame
data = unemployment_data.copy()

# Create the DataFrame
df = pd.DataFrame(data)



# Creating the Sunburst Chart
fig = px.sunburst(df,
                  path=['Region', 'Date', 'Area'],
                  values='Estimated Employed',
                  color='Estimated Unemployment Rate(%)',
                  color_continuous_scale='RdYlBu',
                  hover_data={'Estimated Labour Participation Rate(%)': True})

# Updating the layout
fig.update_layout(
    title="Sunburst Chart of Employment Data by Region and Date",
    coloraxis_colorbar=dict(title="Unemployment Rate(%)"),
    height=600
)

# Show the plot
fig.show()
```
